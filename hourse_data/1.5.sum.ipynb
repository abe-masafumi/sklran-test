## データを追加したい時にdfとdfを結合

```python
import pandas as pd
import time
```

```python
pd.set_option('display.max_rows',5)
pd.options.display.width = 200
pd.options.display.column_space = 100
```

```python
df1 = pd.read_csv("1.2.test1.csv")
df2 = pd.read_csv("1.2.test2.csv")
# df2 = pd.read_pickle("df.pickle")
print(len(df1))
print(len(df2))
```

```python
df1 = df1.iloc[:,:15]
df2 = df2.iloc[:,:15]
```

```python
# pd.set_option('display.max_rows',None)

df = pd.concat(objs=[df1, df2])
# df
```

```python
len(df)
```

```python
df.to_pickle("1.7.df.pickle")
time.sleep(2)
```
