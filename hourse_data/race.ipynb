```python
race_id_lists = ["202209011113"]

# for i in range(1,11):
#   print(i)



def scrape_race_results(race_id_lists):
  race_result_dict = {}
  for race_id in race_id_lists:
    try:
      url = "https://race.netkeiba.com/race/result.html?race_id=" + race_id
      race_result_dict[race_id] = pd.read_html(url)[0]
      print(race_result_dict[race_id])
    except UnicodeDecodeError as e:
      print(e)
      print(f"レースID: {race_id} のデータはありません")
  return race_result_dict
race_result = scrape_race_results(race_id_lists)
race_result
```
